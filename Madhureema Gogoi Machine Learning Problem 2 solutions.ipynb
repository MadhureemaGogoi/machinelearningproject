```python
stopwords = nltk.corpus.stopwords.words('english') +list(string.punctuation)
add_to_stop_words=[',','.','-',';',':','--',"'",'(',')']
stopwords.extend(add_to_stop_words)
stopwords=set(stopwords)
word_list1 = (x.lower() for x in word_list1)
word_list1_clean = [word for word in word_list1 if word not in stopwords]
word_list1_freq = nltk.FreqDist(word_list1_clean)
word_list1_features = [item[0] for item in word_list1_freq.most_common(2000)]
```
